```python
import yfinance as yf
import matplotlib.pyplot as plt

interval = '1d'
period = '1y'
data = yf.download(tickers='GC=F', period=period, interval=interval)
gold_data = data['Adj Close']
gold_date = data.index

data = yf.download(tickers='CL=F', period=period, interval=interval)
oil_data = data['Adj Close']
oil_date = data.index

nifty = yf.download(tickers='^NSEI', period=period, interval=interval)
nifty_data = nifty['Adj Close']
nifty_date = nifty.index
```

```python
# ((data/data.iloc[0]) * 100)
gold_data = (gold_data/gold_data.iloc[0]) * 100
oil_data = (oil_data/oil_data.iloc[0]) * 100
nifty_data = (nifty_data/nifty_data.iloc[0] * 100)
```

```python
plt.plot(gold_date, gold_data, label='gold');
plt.plot(oil_date, oil_data, label='oil');
plt.plot(nifty_date, nifty_data, label='nifty')
plt.legend()
plt.figure(figsize=(25, 20))
plt.show();
```

```python
plt.scatter(gold_date[-1: -50: -1], gold_data[-1: -50: -1], label='gold', alpha=0.5);
plt.scatter(oil_date[-1: -50: -1], oil_data[-1: -50: -1], label='oil', alpha=0.5);
plt.scatter(nifty_date[-1: -50: -1], nifty_data[-1: -50: -1], label='nifty', alpha=0.5)
plt.legend()
plt.figure(figsize=(25, 20))
plt.show();
```
